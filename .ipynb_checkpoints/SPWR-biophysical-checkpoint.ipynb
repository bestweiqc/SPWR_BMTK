# Sharp-Wave Ripples in Basolateral Amygdala
This notebook will create the SPWR model using BMTK.

````python
%%bash
rm -r network/*
````

## 1) Make the cells

````python
from bmtk.builder import NetworkBuilder
import numpy as np
from bmtk.builder.auxi.node_params import positions_cuboid, positions_list

# Initialize our network
net = NetworkBuilder("SPWR_biophysical")

# Create the possible x,y,z coordinates
xside_length = 1400; yside_length = 1400; height = 200; min_dist = 20;
x_grid = np.arange(0,xside_length+min_dist,min_dist)
y_grid = np.arange(0,yside_length+min_dist,min_dist)
z_grid = np.arange(0,height+min_dist,min_dist)
xx, yy, zz = np.meshgrid(x_grid, y_grid, z_grid)
pos_list = np.vstack([xx.ravel(), yy.ravel(), zz.ravel()]).T

# Choose positions for population 1
numPN = 50
numINT = 10
inds = np.random.choice(np.arange(0,np.size(pos_list,0)),numPN,replace=False)

pos = pos_list[inds,:]
# Add a population of 10 nodes (all of which share model_type, dynamics_params, etc.)

net.add_nodes(N=numPN, pop_name='Pyr',
              positions=positions_list(positions=pos),
              mem_potential='e',
              model_type='biophysical',
              model_template='ctdb:Biophys1.hoc',
              model_processing='aibs_perisomatic',
              dynamics_params='472363762_fit.json',
              morphology='Scnn1a_473845048_m.swc')

# Get rid of coordinates already used
pos_list = np.delete(pos_list,inds,0)

# Pick new coordinates
inds = np.random.choice(np.arange(0,np.size(pos_list,0)),numINT,replace=False)
pos = pos_list[inds,:]

net.add_nodes(N=numINT, pop_name='PV',
              positions=positions_list(positions=pos),
              mem_potential='e',
              model_type='biophysical',
              model_template='ctdb:Biophys1.hoc',
              model_processing='aibs_perisomatic',
              dynamics_params='472363762_fit.json',
              morphology='Scnn1a_473845048_m.swc')
````

We will create an external network of inputs

````python
thalamus = NetworkBuilder('mthalamus')
thalamus.add_nodes(N=10,
                   pop_name='tON',
                   potential='exc',
                   model_type='virtual')
````

## 2) Connect the cells

````python
from bmtk.builder.auxi.edge_connectors import distance_connector, connect_random

# Create connections between Pyr --> PV cells
net.add_edges(source={'pop_name': 'Pyr'}, target={'pop_name': 'PV'},
              connection_rule=distance_connector,
              connection_params={'d_weight_min': 0.0, 'd_weight_max': 0.34, 'd_max': 500.0, 'nsyn_min': 0, 'nsyn_max': 10},
              syn_weight=5.0e-03,
              dynamics_params='AMPA_ExcToExc.json',
              model_template='Exp2Syn',
              distance_range=[0.0, 300.0],
              target_sections=['somatic'],
              delay=2.0)


net.build()
net.save_nodes(output_dir='network')
net.save_edges(output_dir='network')


# Create connections between "thalamus" and Pyramidals
thalamus.add_edges(source=thalamus.nodes(), target=net.nodes(pop_name='Pyr'),
                   connection_rule=connect_random,
                   connection_params={'nsyn_min': 0, 'nsyn_max': 12},
                   syn_weight=5.0e-04,
                   target_sections=['somatic'],
                   delay=2.0,
                   distance_range=[0.0, 300.0],
                   dynamics_params='AMPA_ExcToExc.json',
                   model_template='Exp2Syn')


# Build and save our network

thalamus.build()
thalamus.save_nodes(output_dir='network')
thalamus.save_edges(output_dir='network')
````

Set up the files for simulation, using command line:

```bash
 $ python -m bmtk.utils.sim_setup -n network --membrane_report-vars v,cai --membrane_report-cells 10,11 --membrane_report-sections soma --tstop 1000.0 --dt 0.1 bionet
```

You can also run the cell below to call the build_env_bionet function.

Also our cortex cell uses a Scnn1a model we can download from the Allen Cell-Types Database
```bash
 $ wget http://celltypes.brain-map.org/neuronal_model/download/482934212
 $ unzip 482934212
 $ cp fit_parameters.json biophys_components/biophysical_neuron_templates/472363762_fit.json
 $ cp reconstruction.swc biophys_components/morphologies/Scnn1a_473845048_m.swc
```

````python
from bmtk.utils.sim_setup import build_env_bionet
build_env_bionet(network_dir='network', tstop=2000.0, dt=0.1,
                 reports={'membrane_report': {
                         "module": "membrane_report",
                         'cells': 'all',
                         'variable_name': ['cai', 'v'],
                         "file_name": "cell_vars.h5",
                         "sections": "soma",
                     }})
````

Generate spike trains:

````python
from bmtk.utils.spike_trains import SpikesGenerator

sg = SpikesGenerator(nodes='network/mthalamus_nodes.h5', t_max=3.0)
sg.set_rate(15.0)
sg.save_csv('thalamus_spikes.csv', in_ms=True)
````

Edit the simulation_config.json file thusly:

```json
"inputs": {
    "tc_spikes": {
      "input_type": "spikes",
      "module": "csv",
      "input_file": "${BASE_DIR}/thalamus_spikes.csv",
      "node_set": "mthalamus"
    }
}
```

## 3) Run the simulation
This can be done through the command line:
```bash
$ python run_bionet.py simulation_config.json
```


## 4) Inspect Network / Analyse Results

````python
%matplotlib notebook
from bmtk.analyzer import nodes_table
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pdb
from mpl_toolkits.mplot3d import Axes3D

A = nodes_table(nodes_file='network/SPWR_biophysical_nodes.h5', population='SPWR_biophysical')

dset = np.array(A['positions'].values)
pos_df = pd.DataFrame(list(dset))
pos_ds = pos_df.values


fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(pos_ds[0:599,0],pos_ds[0:599,1],pos_ds[0:599,2],color='red')
ax.scatter(pos_ds[600:699,0],pos_ds[600:699,1],pos_ds[600:699,2],color='blue')
````

````python
from bmtk.utils.cell_vars import CellVarsFile
from bmtk.analyzer.cell_vars import _get_cell_report
import matplotlib.pyplot as plt

config_file = "simulation_config.json"
report_name = None
report_name, report_file = _get_cell_report(config_file, report_name)

var_report = CellVarsFile(report_file)
time_steps = var_report.time_trace

plt.plot(time_steps, var_report.data(gid=10,var_name='v'))
plt.plot(time_steps, var_report.data(gid=11,var_name='v'))
````
