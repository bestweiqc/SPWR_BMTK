```python
cell_of_interest = 100

# Get connections
sources = edges_info['edges']['SPWR_biophysical_to_SPWR_biophysical']['source_node_id'].value
targets = edges_info['edges']['SPWR_biophysical_to_SPWR_biophysical']['target_node_id'].value
connections = pd.DataFrame({'source':sources[sources==cell_of_interest], 'target':targets[sources==cell_of_interest]})

# Build a graph
G = nx.from_pandas_edgelist(connections, edge_attr=True)

# nodelist - nodes to plot
# colorlist - colors of those nodes
nodelist = []
colorlist = []
nodelist.append(cell_of_interest)
if cell_of_interest < num_PN:
    colorlist.append('#FF0000')
elif cell_of_interest < num_PV:
    colorlist.append('#00FF00')
else:
    colorlist.append('#0000FF')
for i in np.arange(0,connections.shape[0]):
    nodelist.append(connections['target'].iloc[i])
    if connections['target'].iloc[i] < num_PN:
        colorlist.append('#FF0000')
    elif connections['target'].iloc[i] < num_PV:
        colorlist.append('#00FF00')
    else:
        colorlist.append('#0000FF')

fig3 = plt.figure(figsize=(7,5))
nx.draw_networkx(G,pos,with_labels=False,node_size=50,nodelist=nodelist,node_color=colorlist)
plt.show()
```
